```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

latitudes = [grid[i][0] for i in range(len(grid))]
longitudes = [grid[i][1] for i in range(len(grid))]
plt.scatter(latitudes, longitudes)
```
